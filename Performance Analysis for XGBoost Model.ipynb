```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create some example data
categories = ['Accuracy', 'Precision', 'Recall', 'F1_Score']
values = [100,100,100,100]

custom_colors = ['salmon','steelblue', 'seagreen', 'mediumpurple']


# Set up the figure and axis
plt.figure(figsize=(8, 6))

# Create the bar plot
sns.barplot(x=categories, y=values,palette=custom_colors,edgecolor='black')

# Set labels and title
plt.xlabel('Factors')
plt.ylabel('Performance')
plt.title('Performance Analysis for XGBoost Model')

# Show the plot
plt.show()
```
